```python
def resize_list(X):
    img_list = []
    for image in X:
        row = image.shape[0]
        col = image.shape[1]
        row_low = row//2 - 150
        row_high = row//2 + 150
        col_low = col//2 - 125
        col_high = col//2 + 125
        new_image = image[row_low:row_high, col_low:col_high, :]
        img_list.append(new_image)
    return np.array(img_list)
```
